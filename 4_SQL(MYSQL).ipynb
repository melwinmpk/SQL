# calucalting the week and its rating 

select 
x.Driver,x.week_no,
case when x.cnt <= 3 then 3
     when x.cnt >= 5 then 5
     else x.cnt end as rating
from ( select Driver, WEEK(Date_OF_TRIP) as week_no,count(*) as cnt 
       from TRIP_TABLE group by Driver,WEEK(Date_OF_TRIP)) x;

with temp as (select 
x.Driver,x.week_no,
case when x.cnt <= 3 then 3
     when x.cnt >= 5 then 5
     else x.cnt end as rating
from ( select Driver, WEEK(Date_OF_TRIP) as week_no,count(*) as cnt 
       from TRIP_TABLE group by Driver,WEEK(Date_OF_TRIP)) x)
select t1.Driver from temp t1,temp t2 
where 
t1.Driver = t2.Driver and
t1.week_no > t2.week_no and 
t1.rating > t2.rating;

mysql> WITH RECURSIVE
    -> repeats (col,count,n) as
    -> (
